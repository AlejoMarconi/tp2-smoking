## Duplicados

Hay duplicados a nivel features (sin contar `ID`), pero **`ID` es PK** (50000 valores unicos sobre 50000 filas). Siguiendo el criterio discutido en clase, **no se eliminan** porque (a) cada `ID` es un caso distinto y (b) el modelo en produccion no podra validar a priori que un nuevo registro sea "duplicado".

```python
print('Duplicados (todas las cols)        :', df.duplicated().sum())
print('Duplicados (sin ID, todas iguales) :', df.drop(columns=["ID"]).duplicated().sum())
print('IDs unicos                         :', df['ID'].nunique(), '/', len(df))
```

### Variables a marcar para preprocesamiento

- **`oral`** tiene **un unico valor** (`'Y'`) en el dataset etiquetado y tambien en el sin etiquetar. No aporta informacion y, en ensambles tipo Random Forest, agregar features constantes puede perjudicar la calidad del splitting. Se elimina en el pipeline de preprocesamiento.
- `hearing(left)`, `hearing(right)` y `dental caries` son binarias.
- `gender` y `tartar` son categoricas string a codificar.
- El resto son numericas con dispersion suficiente para el modelo.

```python
print('oral (etiquetado):    ', df['oral'].unique())
print('oral (sin etiquetar): ', df_unlabeled['oral'].unique())
```

## Target (`smoking`)

Distribucion ~63% no fumadores / ~37% fumadores. Es un desbalance moderado, no extremo, asi que **no** se aplica re-muestreo agresivo; se usa `stratify` al hacer el split y `class_weight='balanced'` en los modelos que lo soportan.

```python
counts = df['smoking'].value_counts().sort_index()
prop = df['smoking'].value_counts(normalize=True).sort_index()
pd.DataFrame({'count': counts, 'prop': prop.round(4)})
```

## Coherencia de columnas entre los dos datasets

El dataset sin etiquetar **debe** tener exactamente las mismas columnas que el etiquetado salvo `smoking`.

```python
only_in_labeled = set(df.columns) - set(df_unlabeled.columns)
only_in_unlabeled = set(df_unlabeled.columns) - set(df.columns)
print('Solo en etiquetado    :', only_in_labeled)
print('Solo en sin etiquetar :', only_in_unlabeled)
```

> Diferencia esperada: solo `smoking` esta en el etiquetado. Confirmado.
